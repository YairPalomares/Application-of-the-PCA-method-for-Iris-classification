```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

data = pd.read_csv("iris.csv")
data
```

```python
#Preprocesado y modelado
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from numpy import linalg as LA
```

```python
#Determinar el número de renglones de la base de datos
index = data.index
renglones = len(index)

#Estandarización de las variables
df2 = data.drop("variety", axis=1)
df2 = StandardScaler().fit_transform(df2)
df2
```

```python
#Calcular la matriz de correlaciones para la matriz transformada
A = (1/renglones) * np.dot(df2.T, df2)
A
```

```python
#Entrenamiento del modelo PCA con escalado de los datos
pca_pipe = make_pipeline(PCA())
pca_pipe.fit(df2)

#Se extrae el modelo entrenado del pipeline
modelo_pca = pca_pipe.named_steps["pca"]

print("Eigenvalores:") #indicadores de explicación de varianza
results = LA.eigvals(A)
print(results)
```

```python
#Porcentaje de la varianza explicada por cada nuevo componente
print("Porcentaje de varianza explicada por componente")
print(modelo_pca.explained_variance_ratio_)
```

```python
#Grado de explicación de la varianza al utilizar los dos primeros componentes
print("Al utilizar los dos primeros componentes se puede explicar un ", (0.7296 + 0.2285)*100, "% de la varianza.")
```

```python
# Cálculo de los eigenvectores
print("Eigenvectores (por renglón):")
pd.DataFrame(data = modelo_pca.components_,
             columns = ["sepal.length", "sepal.width", "petal.length", "petal.width"],
             index = ["PC1", "PC2", "PC3", "PC4"])
```

```python
#Proyecciones de los componentes
proyecciones = np.dot(modelo_pca.components_, df2.T)
proyecciones = pd.DataFrame(proyecciones, index = ["PC1", "PC2", "PC3", "PC4"])
proyecciones = proyecciones.transpose().set_index(data.index)
proyecciones
```

```python
# Mapa de observaciones
proyecciones = pd.concat([proyecciones, data["variety"]], axis=1)

x_setosa = proyecciones[proyecciones["variety"] == "Setosa"].iloc[:,0]
y_setosa = proyecciones[proyecciones["variety"] == "Setosa"].iloc[:,1]
x_setosa = x_setosa.to_numpy()
y_setosa = y_setosa.to_numpy()

x_versicolor = proyecciones[proyecciones["variety"] == "Versicolor"].iloc[:,0]
y_versicolor = proyecciones[proyecciones["variety"] == "Versicolor"].iloc[:,1]
x_versicolor = x_versicolor.to_numpy()
y_versicolor = y_versicolor.to_numpy()

x_virginica = proyecciones[proyecciones["variety"] == "Virginica"].iloc[:,0]
y_virginica = proyecciones[proyecciones["variety"] == "Virginica"].iloc[:,1]
x_virginica = x_virginica.to_numpy()
y_virginica = y_virginica.to_numpy()

fig, ax = plt.subplots()
ax.set_title("Mapa de Observaciones")
ax.scatter(x_setosa, y_setosa, color = "blue", label = "Setosa")
ax.scatter(x_versicolor, y_versicolor, color = "red", label = "Versicolor")
ax.scatter(x_virginica, y_virginica, color = "green", label = "Virginica")
plt.legend()
plt.show()
```

```python
componentes_2 = pd.DataFrame(data = modelo_pca.components_,
                             columns = ["sepal.length", "sepal.width", "petal.length", "petal.width"],
                             index = ["PC1", "PC2", "PC3", "PC4"])
componentes_2 = componentes_2.iloc[0:2, :]
componentes_2 = componentes_2.T
componentes_2
```

```python
# Mapa de factores
x = componentes_2.iloc[:,0]
y = componentes_2.iloc[:,1]
z = componentes_2.index
x = x.to_numpy()
y = y.to_numpy()

fig, ax = plt.subplots()
ax.set_title("Mapa de Factores")
ax.scatter(x, y)

for i, txt in enumerate(z):
    ax.annotate(txt, (x[i], y[i]))
```

```python
# ¿Qué puede concluir al compararlos? 
print("Posterior al análisis PCA y la graficación de los mapas de observaciones y factores, se puede concluír que el factor característico de las setosas es el ancho del sépalo, ya que como se puede observar en los datos, este suele ser más grande que el resto de las especies de flores analizadas.")
print()
print("Por otro lado, también se puede concluir que la longitud del sépalo en las virginias es mayor a los otros pétalos estudiados.")
print()
print("Finalmente, otra de las deducciones que nos permite realizar el presente análisis de clasificación, es que los pétalos versicolores son los que presentan mayor tamaño tanto en ancho como largo.")
```
